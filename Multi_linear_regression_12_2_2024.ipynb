```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/1000_Companies.csv')
df
```

```python
X=df.drop('Profit',axis='columns')
y=df['Profit']
```

```python
X
```

```python
y
```

```python
# Step 3:Find the correlation using Heatmap
```

```python
sns.heatmap(df.corr())
```

```python
# Step 4 & 5: Identify categorical Data and encoding
```

```python
dummies = pd.get_dummies(df['State'])
dummies
```

```python
merged = pd.concat([X,dummies],axis='columns')
merged
```

```python
X=merged.drop(['State','California'],axis='columns')
X
```

```python
# Step 6:Split the data
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
# TRAINING THE MODEL
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
model.fit(X_train,y_train)
```

```python
#Step 8:Predict the values
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_test
```

```python
# Step 9:Model Metrix calculation
```

```python
model.score(X_test,y_test)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/insurance_data.csv')
df
```

```python
plt.scatter(df.age,df.bought_insurance,marker='+',color='blue')
```

```python
#training and testing split
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['age']],df.bought_insurance, train_size = 0.7)
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_test
```

```python
model.score(X_test,y_test)
```
