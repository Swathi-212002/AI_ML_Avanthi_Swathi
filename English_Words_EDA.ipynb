Change the column name 'A' to the words

```python
df = df.rename(columns={'A':'Names'})
```

```python
df
```

Add character count of each word as another column in the dataframe

```python
df['Names']=df['Names'].fillna('')
df['Character Count']=df['Names'].apply(len)
df
```

```python
df
```

Add a column called value which is the sum of the numerical positions of the alphabets in the word
(Ex: aah = 1 + 1 + 8 = 10)

```python
def summing(word):
  sum=0
  word.lower()
  for char in word:
    position=ord(char)-96
    sum+=position
  return sum

df['Values']=df['Names'].apply(summing)
```
